```python
#plot each feature 
import csv
test_data = []
with open(r'C:/Users/asfahad/Documents/Big Data London 2013/AdjointFashematians/data/test.csv', 'rb') as csvfile:
    spamreader = csv.DictReader(csvfile)
    for row in spamreader:
        test_data.append(row)
```

```python
" - ".join(test_data[0].keys())
k = test_data[0].keys()
#get all unique keys per person
s = set()
for i in k:
    s.add(i[2:])
print s
print test_data[0]
```

```python
#[peaks([x,y]) for x,y in zip(xscat,yscat)]

all_diffs = {}
for feature in s:
    all_diffs[feature] = []
    for entry in test_data:
        person_a = 'A_' + feature
        person_b = 'B_' + feature
        all_diffs[feature].append(numpy.log(float(entry[person_a])+1) - numpy.log(float(entry[person_b])+1)  )
```
